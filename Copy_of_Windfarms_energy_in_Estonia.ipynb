```python
# I decided to study some parameters of the wind which might have affect on the wind power generation
# and electricity production.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# First, I downloaded from the  ttps://globalwindatlas.info/en/area/Estonia/Harju  website the data
# on Mean Wind Speed over 50 indexes in  Estonia in CSV format and uploaded here
```

```python
WindSpeedmean=pd.read_csv("WindSpeed mean over Estonia.csv")
WindSpeedmean
```

```python
# Then I uploaded from the same website the data on 50 indexes on Mean Power Density over Estonia
# which means Power density in Megawatts , the ratio of the windfarm's rated capacity to its ground area.
```

```python
PowerDensitymean=pd.read_csv("powerDensity mean over Estonia.csv")
PowerDensitymean
```

```python
# Then I decided to visualise the data of the mean Wind Speed against the mean Power Density
```

```python
import numpy as np
import matplotlib.pyplot as plt



x = np.array(WindSpeedmean.val)
y = np.array(PowerDensitymean.val)

plt.plot(x, y)

plt.xlabel("Wind Speed mean")
plt.ylabel("Power Density mean")

plt.show()
```
